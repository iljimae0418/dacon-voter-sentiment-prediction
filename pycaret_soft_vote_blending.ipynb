```python
train = pd.read_csv('train.csv') 
test = pd.read_csv('test_x.csv') 
submission = pd.read_csv('sample_submission.csv') 
```

```python
train = train.drop(columns=['index']) 
test = test.drop(columns=['index'])
```

```python
from pycaret.classification import *
```

```python
clf = setup(data = train, target = 'voted')
```

After completing the setup, the compare_models() function trains all the models in teh model library and scores them using stratified cross validation for metric evaluation. The output prints a score grid that shows average accuracy, AUC, Recall, Precision, F1, Kappa and MCC across the folds (10 by default) along with training times.  

```python
best_7 = compare_models(sort = 'AUC', n_select = 7)
```

```python
blended = blend_models(estimator_list = best_7, fold = 5, method = 'soft')
```

```python
pred_holdout = predict_model(blended)
```

```python
final_model = finalize_model(blended)
```

```python
predictions = predict_model(final_model, data = test)
```

```python
submission['voted'] = predictions['Score']
```
